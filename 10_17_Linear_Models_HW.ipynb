```python
import pandas as pd

car_data_raw = pd.read_csv("https://cdn.c18l.org/vehicles_lab.csv")
```

```python
car_data_raw.head()
```

Part 1 - Feature Selection

```python
new_data = car_data_raw.loc[:,["price", "year", "manufacturer", "condition", "odometer", "posting_date"]]
```

```python
new_data.head()
new_data.describe()
```

Part 2 - Data Cleaning

```python
#create variables for average price and standard deviation price

avg_price = new_data['price'].mean()
std_price = new_data['price'].std()
```

```python
#normalize price with mean squared errors equation
new_data["price"] = (new_data["price"]- avg_price) / std_price
```

Part 3 - Feature Engineering

```python
new_data["Old_and_Cheap"] = (new_data["price"] <5000) & (new_data["year"] < 1990)
#wanted to add a column that has old and cheap cars
```

```python
new_data["Ford_Excellent"] = (new_data["manufacturer"] == "ford") & (new_data["condition"] == 'excellent')
#wanted to see which cars are made by ford and are excellent
```

```python
new_data.head()
```

Part 4 - Multinomial Classification

```python
new_data = pd.cut(new_data['price'], bins=range(0, int(new_data['price'].max()) + 10000, 10000), right=False, include_lowest=True)
```
